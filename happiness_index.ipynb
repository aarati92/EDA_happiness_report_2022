```python
#First we import all the required libraries which will assist in our EDA.

import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# Next we import our csv file that is the "world happiness report 2022".

data = pd.read_csv("happiness_report_2022.csv")
```

```python
#To see if any coutry name has been repeated:
data['Country'].value_counts()
```

```python
#Using numpy to see what is the highest happiness score.

data['Happiness score'].argmax()
```

```python
# To see the details of a particular row.
# when displaying one single entry from the dataFrame, the values won't appear as row but will be displayed as a column.

data.iloc[data['Happiness score'].argmax()]
```

```python
# To see basic statistical features of the data.

data.describe()
```

From above table we can see that the happiness score ranges from 2.40 to 7.82 .

```python
#to see top 5 countries with highest gdp

sns.barplot(x='Country', 
            y='Explained by: GDP per capita',
            data=data.nlargest(5, 'Explained by: GDP per capita'))
plt.show()
```

```python
#relationship between gdp and happiness:

plt.figure(figsize=(10,5))
sns.scatterplot(x=data['RANK'],y=data['Explained by: GDP per capita'])
plt.show()
```

```python
columns = ['Happiness score',	'Explained by: GDP per capita', 'Explained by: Social support', 'Explained by: Healthy life expectancy']
sns.pairplot(data[columns])
```

We can also perform bivariate analysis - we can look at pairs of two variables to explore a possible relation between them.
In bivariate analysis, we look at scatterplots like above and search for a pattern like straight/sloped/linear/inverted line. This is a visual representation of two variables that correlate.
Diagonally, from upper left to lower right are the histograms we've made in the univariate analysis.

From the above plots we can come to a basic conclusion that happiness score increases as the other factors like gdp/social support/health life expectancy increases.
In these scatter plots we are comparing 2 variables,hence this is a bivariate analysis.
